# Load some house sales data

# Exploring the data for housing sales 

The house price is correlated with the number of square feet of living space.

```python
graphlab.canvas.set_target('ipynb')
sales.show(view="Scatter Plot", x="sqft_living", y="price")
```

# Create a simple regression model of sqft_living to price

Split data into training and testing.  
We use seed=0 so that everyone running this notebook gets the same results.  In practice, you may set a random seed (or let GraphLab Create pick a random seed for you).  

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

## Build the regression model using only sqft_living as a feature

```python
sqft_model = graphlab.linear_regression.create(train_data, target='price', features=['sqft_living'],validation_set=None)
```

# Evaluate the simple model

```python
print test_data['price'].mean()
```

```python
print sqft_model.evaluate(test_data)
```

RMSE of about \$255,170!

### Comparision between SFrame, sklearn and Statsmodel

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import *

def evaluate(X_train, X_test, y_train, y_test):
    sk_model = LinearRegression(copy_X=True)
    sk_model.fit(X_train, y_train)
    
    N = len(y_test)
    rmse = np.sqrt(((y_test - sk_model.predict(X_test)) ** 2).sum() / N)
    print('RMSE: %s' % rmse)
    return rmse

import statsmodels.api as sm

def evaluate_sm(X_train, X_test, y_train, y_test):
    xtrain = X_train.convert_objects(convert_numeric=True)
    ytrain = y_train.convert_objects(convert_numeric=True)
    xtest = X_test.convert_objects(convert_numeric=True)
    ytest = y_test.convert_objects(convert_numeric=True)
    sm_model = sm.OLS(ytrain, xtrain)
    sm_result = sm_model.fit()
    rmse = np.sqrt(mean_squared_error(ytest, sm_result.predict(xtest)))
    print('RMSE_SM: %s' % rmse)
    return rmse

def evaluate_glsar(X_train, X_test, y_train, y_test):
    xtrain = X_train.convert_objects(convert_numeric=True)
    ytrain = y_train.convert_objects(convert_numeric=True)
    xtest = X_test.convert_objects(convert_numeric=True)
    ytest = y_test.convert_objects(convert_numeric=True)
    sm_model = sm.GLSAR(ytrain, xtrain)
    sm_result = sm_model.fit()
    rmse = np.sqrt(mean_squared_error(ytest, sm_result.predict(xtest)))
    print('RMSE_GLSAR: %s' % rmse)
    return rmse

def evaluate_wls(X_train, X_test, y_train, y_test):
    xtrain = X_train.convert_objects(convert_numeric=True)
    ytrain = y_train.convert_objects(convert_numeric=True)
    xtest = X_test.convert_objects(convert_numeric=True)
    ytest = y_test.convert_objects(convert_numeric=True)
    sm_model = sm.WLS(ytrain, xtrain)
    sm_result = sm_model.fit()
    rmse = np.sqrt(mean_squared_error(ytest, sm_result.predict(xtest)))
    print('RMSE_WLS: %s' % rmse)
    return rmse


rmse1 = evaluate(train_data.to_dataframe()[my_features],
                 test_data.to_dataframe()[my_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)
rmse2 = evaluate(train_data.to_dataframe()[advanced_features],
                 test_data.to_dataframe()[advanced_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)

rmse1_sm = evaluate_sm(train_data.to_dataframe()[my_features],
                 test_data.to_dataframe()[my_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)
rmse2_sm = evaluate_sm(train_data.to_dataframe()[advanced_features],
                 test_data.to_dataframe()[advanced_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)

rmse1_glsar = evaluate_glsar(train_data.to_dataframe()[my_features],
                 test_data.to_dataframe()[my_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)
rmse2_glsar = evaluate_glsar(train_data.to_dataframe()[advanced_features],
                 test_data.to_dataframe()[advanced_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)

rmse1_wls = evaluate_wls(train_data.to_dataframe()[my_features],
                 test_data.to_dataframe()[my_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)
rmse2_wls = evaluate_wls(train_data.to_dataframe()[advanced_features],
                 test_data.to_dataframe()[advanced_features],
                 train_data.to_dataframe().price,
                 test_data.to_dataframe().price)

print my_features_model.evaluate(test_data)['rmse']
print advanced_features_model.evaluate(test_data)['rmse']
print('%s - %s = %s' % (rmse1, rmse2, rmse1-rmse2))
print('%s - %s = %s' % (rmse1_sm, rmse2_sm, rmse1_sm-rmse2_sm))
print('%s - %s = %s' % (rmse1_glsar, rmse2_glsar, rmse1_glsar-rmse2_glsar))
print('%s - %s = %s' % (rmse1_wls, rmse2_wls, rmse1_wls-rmse2_wls))
```
